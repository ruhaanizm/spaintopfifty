## Normalizing Song and artist naming

```python
df['song'] = df['song'].str.strip()
df['artist'] = df['artist'].str.strip()
```

```python
#creating a unique song id to identify individual songs

df['song_id'] = df['song'] + ' by ' + df['artist']
```

```python

first_appearance = df.groupby('song_id')['date'].min().reset_index()
first_appearance.rename(columns={'date': 'entry_date'}, inplace=True)
```

```python
df = df.merge(first_appearance, on='song_id', how='left')
```

```python
df[['song', 'artist', 'date', 'entry_date']].head()
```

```python
df.head()
```

```python
song_group = df.groupby('song_id')
```

```python
lifecycle_df = song_group.agg(
    entry_date=('date', 'min'),
    exit_date=('date', 'max')
).reset_index()
```

```python
lifecycle_df['total_days'] = (
    lifecycle_df['exit_date'] - lifecycle_df['entry_date']
).dt.days + 1

#(+1 because same day = 1 day, not 0)
```

```python
peak_pos = song_group['position'].min().reset_index()
peak_pos.rename(columns={'position': 'peak_position'}, inplace=True)

lifecycle_df = lifecycle_df.merge(peak_pos, on='song_id')
```

```python
# time to peak

peak_dates = df.loc[df.groupby('song_id')['position'].idxmin()][['song_id', 'date']]
peak_dates.rename(columns={'date': 'peak_date'}, inplace=True)

lifecycle_df = lifecycle_df.merge(peak_dates, on='song_id')

lifecycle_df['time_to_peak'] = (
    lifecycle_df['peak_date'] - lifecycle_df['entry_date']
).dt.days
```

Lifecycle Stage Classification
Classify songs into stages:

• New Entry (≤7 days)
• Growth Phase (rank improving)
• Peak Phase (Top 10 stability)
• Mature Phase (stable mid-rank)
• Decline Phase (rank deterioration)

```python
df['days_since_entry'] = (df['date'] - df['entry_date']).dt.days
```

```python
df = df.sort_values(['song_id', 'date'])
```

```python
df.head()
```

```python
df['prev_position'] = df.groupby('song_id')['position'].shift(1)
df['rank_change'] = df['prev_position'] - df['position']
```

```python
df.head()
```

```python
df['stage'] = 'Other'  # default

df.loc[df['days_since_entry'] <= 7, 'stage'] = 'New Entry'
```

```python
# if song in top 10

df.loc[df['position'] <= 10, 'stage'] = 'Peak'
```

```python
# if song is growing

df.loc[
    (df['rank_change'] > 0) & (df['stage'] == 'Other'),
    'stage'
] = 'Growth'
```

```python
# if song is declining
df.loc[
    (df['rank_change'] < 0) & (df['stage'] == 'Other'),
    'stage'
] = 'Decline'
```

```python
# song reached maturity

df.loc[df['stage'] == 'Other', 'stage'] = 'Mature'
```

```python
df['stage'].value_counts()
```

Playlist Rotation & Churn Analysis
• Daily entry and exit counts
• Average churn rate
• Stability vs volatility measurement
• Rotation comparison across months

```python
daily_songs = df.groupby('date')['song_id'].apply(set).reset_index()
```

```python
daily_songs
```

```python
daily_songs = daily_songs.sort_values('date')
daily_songs
```

```python

#entries = today - yesterday
#exits   = yesterday - today

entries = []
exits = []

for i in range(len(daily_songs)):
    if i == 0:
        entries.append(0)
        exits.append(0)
    else:
        prev_set = daily_songs.loc[i-1, 'song_id']
        curr_set = daily_songs.loc[i, 'song_id']
        
        new_entries = len(curr_set - prev_set)
        new_exits = len(prev_set - curr_set)
        
        entries.append(new_entries)
        exits.append(new_exits)

daily_songs['entries'] = entries
daily_songs['exits'] = exits
```

```python
daily_songs
```

```python
daily_songs['churn_rate'] = daily_songs['entries'] / 50

# playlist size = 50
```

```python
daily_songs
```

```python

avg_churn = daily_songs['churn_rate'].mean()
avg_churn
```

0.03==3.4% of 50 ≈ 1.7 songs/day
1-2 songs change per day, very stable playlist

```python
daily_songs['month'] = daily_songs['date'].dt.to_period('M')

monthly_churn = daily_songs.groupby('month')['churn_rate'].mean().reset_index()
daily_songs
```

```python
monthly_churn
```

```python
import plotly.express as px

fig = px.line(daily_songs, x='date', y='churn_rate', title='Daily Churn Rate')
fig.show()
```

```python
monthly_churn['month'] = monthly_churn['month'].astype(str)
```

```python
px.bar(monthly_churn, x='month', y='churn_rate', title='Monthly Churn Comparison')
```

Key Insights and Interpretation
1. Playlist Longevity and Lifecycle Behavior

The analysis shows that songs remain on the playlist for an average of 57.44 days, indicating a relatively long content lifecycle.

Additionally, the average entry-to-peak time of 9.48 days suggests that songs do not achieve peak popularity instantly but instead follow a gradual growth trajectory before reaching their highest rank.

Interpretation:

The Spanish music market favors progressive discovery over instant virality
Songs tend to build momentum over time rather than peak immediately
2. Playlist Stability and Rotation Dynamics

The playlist exhibits an average churn rate of 3.43%, meaning roughly 1–2 songs are replaced daily.

Interpretation:

The playlist is highly stable
Popular songs tend to retain positions for extended periods
Entry barriers for new songs are relatively high
3. Retention Stability Index (Content Durability)

The retention stability index (standard deviation) is 89.97, indicating significant variability in song lifecycles.

Interpretation:

There is high variation in how long songs stay on the playlist
Some songs exit very quickly, while others remain for extended durations
This reflects a hit-driven market where a small number of tracks dominate
4. Explicit vs Non-Explicit Content Performance
Non-explicit songs: 64.05 days
Explicit songs: 61.65 days

Interpretation:

Non-explicit songs show slightly higher retention (approximately 4% longer)
This suggests broader audience appeal and better long-term replayability
However, the difference is moderate, indicating that explicit content still performs competitively
5. Single vs Album Strategy
Single vs Album longevity ratio: 1.22

Interpretation:

Singles remain on playlists approximately 22% longer than album tracks
Focused release strategies are more effective than album-based exposure
Singles appear to be better optimized for playlist performance and retention
6. Album Size vs Lifecycle Stability

Analysis of total tracks per album indicates that smaller track collections tend to perform better in terms of retention, while larger albums show inconsistent results.

Interpretation:

Larger albums dilute listener attention across multiple tracks
Playlist success is driven by individual track strength rather than album size
Singles or smaller releases are more effective in maintaining listener engagement
7. Churn Trends Over Time

Monthly churn analysis shows generally low churn rates (approximately 2–5%), with occasional spikes (e.g., November 2025 reaching approximately 6.7%, and isolated days with significantly higher churn).

Interpretation:

The playlist remains stable over time
Sudden spikes in churn indicate external influences such as major releases, seasonal trends, or industry-driven resets
Overall Strategic Insight

The Spanish Top 50 playlist demonstrates a stable yet hit-driven ecosystem, where a small number of high-performing tracks dominate for extended periods, while the majority experience shorter lifecycles.

Executive Summary

The analysis reveals that Spain’s Top 50 playlist is characterized by high stability and strong content retention, with songs averaging 57.44 days on the chart and low daily churn (approximately 3.4%). Songs typically follow a gradual growth pattern, taking around 9.5 days to reach peak position, indicating a market driven by sustained engagement rather than immediate virality.

Content-wise, non-explicit tracks show slightly better retention, while singles outperform album tracks significantly, remaining on playlists approximately 22% longer. This highlights the effectiveness of focused release strategies over album-based distribution.

Additionally, the high variability in lifecycle durations suggests a hit-driven market structure, where a few dominant tracks maintain long-term presence while others exit rapidly. Periodic spikes in churn further indicate external influences such as major releases or seasonal shifts impacting playlist dynamics.
